# Setups & Get started

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd

recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0], '\n\n\n')
print(recent_grads.head(2), '\n\n\n')
print(recent_grads.tail(2))
```

```python
recent_grads.describe()
```

```python
raw_data_count = len(recent_grads)
```

```python
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads)
print(raw_data_count, cleaned_data_count)
```

# Pandas, Scatter Plots

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter',
                  title='Full-time-job vs Income, correlation',
                  xlim=(0, 300000), ylim=(0, 115000))
```

It seems there isn't any correlation between full-time employees and median salary.

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='Total', y='Median', kind='scatter', 
                  title='Who make more money?',
                 ylim=(20000, 80000), xlim=(0, 80000))
```

It doesn't seems there are any correlation between the popular major and income.

```python
from numpy import arange
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter',
                 title='Do more women make more money?',
                 xlim=(0, 1), ylim=(0, 115000),
                 xticks=0.1*arange(10))
```

As we can see the weak correlation on the plot above, the income decreases while the sharewomen increases.

```python
recent_grads['ShareWomen'].hist(bins=100)
mostly_female = len(recent_grads[recent_grads['ShareWomen']>0.5])
gender_percentage = mostly_female/len(recent_grads['ShareWomen'])
print(gender_percentage)
```

About half of the majors are predominantly male, the other half are predominantly female according to the histogram under bin size of 25, but when increase the bin size to 100, we will see that ShareWomen on upper half seems slightly more than on the lower half.

Conclusionis 55.8% of majors are predominantly female.
